```python
'''START CODE'''

#Import necessary libraries: Numpy,pandas,matplotlib

'''************************Your code here**********************'''
import csv
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
'''************************************************************'''
```

```python
#read meal_info.csv file from provided dataset into a df_meal named variable
#Note: Proper file directory should be provided

'''************************Your code here**********************'''
df_meal = pd.read_csv('../../dataset/assign1/meal_info.csv')
'''************************************************************'''


df_meal.head() #"""This is for displaying first five data points"""
```

```python
#read fulfilment_center_info.csv file from provided dataset
#Note: Proper file directory should be provided

'''************************Your code here**********************'''
df_center = pd.read_csv('../../dataset/assign1/fulfilment_center_info.csv')
'''************************************************************'''

df_center.head() #'''This is for displaying first five data points'''
```

```python
#read train.csv file from provided dataset
#Note: Proper file directory should be provided

'''************************Your code here**********************'''
df_food = pd.read_csv('../../dataset/assign1/train.csv')
'''************************************************************'''

df_food.head()
```

```python
'''Since the provided information is in different files, your work here is to merge them.Look for the functions
in pandas library to do so'''

'''************************Your code here**********************'''
#perform outer join operation
df = pd.merge(pd.merge(df_food,df_meal,on='meal_id',suffixes=('_food','_meal'),how='outer'),df_center,on='center_id',suffixes=('_food','_center'),how='outer')

'''************************************************************'''
df.head(5)
```

```python
'''Here we have used pd.pivot_table() kindly go through the function and mention in comment what it does'''

table = pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)

""" pivot creates a view of existing table for viewing info about some fields(as columns) grouped by some fields(as index) """
""" table stores total num_orders for each category of food """
table.head(5)
```

```python
'''Graph tweaking
************************
Plot a bar graph with title 'Most popular food' for category(x-axis) vs number-of-orders(y-axis)

give x label 'Food items'
give y label 'Quantity sold'


'''

'''************Yourcode*********************'''
category_x = [x for x in table.index]
num_orders_y = [table.loc[x,'num_orders'] for x in category_x]

#bar graph
plt.bar(category_x,num_orders_y)

#xticks 
plt.xticks(rotation=60) #'''Write on comment what you feel this function does'''
"""
-xticks specifies a list of position where the x-ticks (x values)  are to be placed
-with rotation=theta argument, it takes the original position array and roatates by certain angle (y-position changes)
"""
#x-axis labels 
plt.xlabel('Food items') 

#y-axis labels 
plt.ylabel('Quantity sold')

#plot title 
plt.title('Most popular food')

#save plot 
plt.savefig('../../assets/image/assign1_most_popular_food.png')

#display 
plt.tight_layout()
plt.show()

'''************************************************'''
```

```python
'''Comparison of weekly and monthly sales
 Create a new column
* named 'revenue' where each element is product of checkout_price and num_orders 
** named 'month' by using ['week'] column (week column value divided by 4 gives month value)'''



'''**************************Your code******************************'''
total_weeks_in_year = 4*12
month_lookup = ['January','February','March','April','May','June','July','August','September','October','November','December']
revenue = df['checkout_price']*df['num_orders']
month = pd.Series([month_lookup[index] for index in list(((df['week']-[1])%total_weeks_in_year)//4)])
df['month']=month
df['revenue']=revenue
#df['revenue'] = df.apply(lambda x: x.checkout_price*x.num_orders,axis=1) 
'''******************************************************************'''
df.head(8)
```

```python
'''Here we have created two list month and month_order ,
store month number in month list and revenue of each month in month_order'''
#list to store month-wise revenue 
month=[] 
month_order=[] 

'''***********************************Your code***********************'''
#pivot table with total revenue for each month
month_no_pivot = pd.pivot_table(data=df.assign(month_no=((df['week']-1)//4)+1),index='month_no',values='revenue',aggfunc=np.sum)
month = [x for x in month_no_pivot.index]
month_order = [month_no_pivot.loc[x,'revenue'] for x in month]
    
'''*********************************************************************'''
print(month[:5])
print(month_order[:5])
```

```python
'''Here we have created two list week and week_order ,you need to store in them mapping the monthly orders'''    
#list to store week-wise revenue 
week = []
week_order = []
'''***********************************Your code***********************'''

week_revenue_pivot = pd.pivot_table(data=df,index='week',values='revenue',aggfunc=np.sum)
week=[x for x in week_revenue_pivot.index] 
week_order=[x for x in week_revenue_pivot['revenue']] 

'''*********************************************************************'''
print(week[:5])
print(week_order[:5])
```

```python
''' Plot two subplots in the same space : one for weekly revenue and other for monthly revenue.
For weekly : Title(Weekly income),x_label(week),y_label(revenue); similarly for monthly revenue.
'''

'''************Yourcode*********************'''
# matplotlib.pyplot.subplots return an instance of Figure and an array of (or a single) Axes
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(30,7))
axes[0].plot(week,week_order)
axes[0].set_xlabel('week')
axes[0].set_ylabel('revenue')
axes[0].set_title('Weekly income')

axes[1].plot(month,month_order)
axes[1].set_xlabel('month')
axes[1].set_ylabel('revenue')
axes[1].set_title('Monthly income')

plt.savefig('../../assets/image/assign1_weekly_monthly_income.png')
plt.tight_layout()
plt.show()
''' Display the plot'''
```
