```python
dev_x = np.random.randint(25,35,11)
py_dev_x = np.random.randint(25,35,11)
print(dev_x)
print()
print(py_dev_x)
```

```python
dev_y = np.random.randint(12000,100000,11)
py_dev_y = np.random.randint(12000,100000,11)
cpp_dev_y = np.random.randint(34000,131112,11)
print(dev_y)
print()
print(py_dev_y)
print()
print(cpp_dev_y)
```

```python
plt.plot(dev_x,dev_y,color='b',label='All Developers')
#plt.plot(py_dev_x,py_dev_y,'o--k',label='Python Developers')
plt.plot(py_dev_x,py_dev_y,marker='o',linestyle=' ',color='black',label='Python Developers')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.title('median salary (USD) by age')
plt.legend()
#plt.legend(['All Developers','Python Developers'])

#format string
#fnt = '[marker][line][color]'


plt.show()
```

```python
plt.plot(dev_x,dev_y,color='b',label='All Developers')
plt.plot(py_dev_x,py_dev_y,marker='o',linestyle='--',color='black',linewidth=3,label='Python Developers')
plt.plot(dev_x,cpp_dev_y,color='g',label='cpp developer')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.title('median salary (USD) by age')
plt.legend()

plt.grid("true")
plt.tight_layout()
plt.savefig('plot1.png')

plt.show()
```
